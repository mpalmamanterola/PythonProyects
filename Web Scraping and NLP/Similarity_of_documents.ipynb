```python
import pandas as pd
from nltk.tokenize import TreebankWordTokenizer
import re
import nltk
import numpy as np
from collections import Counter
```


Se reúne un corpus de al menos 30 documentos. Se guardan en una base de datos, ordenadamente,
y en formato csv.
Utilizando la codificación bolsa de palabras binaria y una métrica coherente,
se compararan los documentos y se muestran los más similiares. Para esto, se construye un léxico ad-hoc y se utiliza un buen algoritmo de tokenización.


```python
df1 = pd.read_csv('CSV/Machine Learning_MIT')
df2 = pd.read_csv('CSV/Obama_MIT')
df3 = pd.read_csv('CSV/Theorem_MIT')
```

```python
df = pd.concat([df1,df2,df3],ignore_index = True)
df
```

```python
Corpus = [df['Body'][i].lower() for i in range(0,len(df))]
puntuacion = set((',', '.', '--', '-', '!', '?', ':', ';', '``', "\"", '(', ')', '[', ']','\''))
Corpus = [x for x in Corpus if x not in puntuacion]
Corpus[0]
```

```python
stopwords = nltk.corpus.stopwords.words('english')
tokenizer = TreebankWordTokenizer()
lexic=0
Tokens = []
for doc in Corpus:
    lexic += len(doc)
    out = re.sub(r'[^\w\s""]','',doc)
    tokens = tokenizer.tokenize(out)
    tokens = [x for x in tokens if x not in stopwords]
    tokens = sorted(tokens)
    Tokens.append(tokens)
```

```python
corpus = {}

for i, sent in enumerate(Corpus):
    corpus[i] = dict((tok, 1) for tok in sorted(sent.split()))
docs = pd.DataFrame.from_records(corpus).fillna(0).astype(int).T
docs
```

```python
import math

def sim_coseno(vec1, vec2):
    
    dot_prod = vec1.dot(vec2.T)
    
    norm_1 = np.linalg.norm (vec1)
    norm_2 = np.linalg.norm (vec2)
    
    return float(dot_prod / (norm_1 * norm_2))
```

```python
S = np.zeros((len(docs),len(docs)))
S[0,0] = 1
S[len(docs)-1,len(docs)-1] = 1 
a = 0

for count in range(0,len(docs)):
    for i in range (1,len(docs)-1):
        sc = sim_coseno(docs.iloc[count] , docs.iloc[i])
    
        if sc >= 0.98 and i != count:
            a += 1
        S[count,i] = sc
            
print('Cantidad de documentos idénticos: '+str(a)+' de '+str(len(docs)))
print(np.triu(S))
```

Ahora, con el mismo corpus anterior, se vectorizan los documentos con el
método TF-IDF y se realiza una comparación entre los documentos. finalmente, se muestran los
documentos más similares en relación a la similitud coseno.

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(min_df = 0.1, encoding='utf-8')
model = vectorizer.fit_transform(Corpus)
print(len(model.todense()))
model.todense().shape
```

```python
S = np.zeros((len(model.todense()),len(model.todense())))
a = 0
S[0,0] = 1
S[44,44] = 1
for count in range(0,len(model.todense())):
    for i in range (1,len(model.todense())-1):
        sc = sim_coseno(model.todense()[count] , model.todense()[i])
    
        if sc >= 0.98 and i != count:
            a += 1
        S[count,i] = sc
print('Cantidad de documentos idénticos: '+str(a)+' de '+str(len(model.todense()) ))
print(np.triu(S))
```
